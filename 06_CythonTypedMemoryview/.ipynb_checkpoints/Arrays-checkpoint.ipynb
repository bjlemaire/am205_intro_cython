```python
%load_ext cython
import numpy as np
```

```python
def py_total(arr):
    I, i = arr.shape[0], 0
    J, j = arr.shape[1], 0
    K, k = arr.shape[2], 0
    total = 0
    for i in range(I):
        for j in range(J):
            for k in range(K):
                total += arr[i, j, k]
    return total
```

```python
%%cython -a
def cy1_total(arr):
    cdef unsigned I = arr.shape[0], i=0
    cdef unsigned J = arr.shape[1], j=0
    cdef unsigned K = arr.shape[2], k=0
    cdef int total = 0
    for i in range(I):
        for j in range(J):
            for k in range(K):
                total += arr[i, j, k]
    return total
```

```python
%%cython -a
def cy2_total(int[:,:,:] arr):
    cdef unsigned I = arr.shape[0], i=0
    cdef unsigned J = arr.shape[1], j=0
    cdef unsigned K = arr.shape[2], k=0
    cdef int total = 0
    for i in range(I):
        for j in range(J):
            for k in range(K):
                total += arr[i, j, k]
    return total
```

```python
%%cython -a
cpdef int cy3_total(int[:,:,:] arr):
    cdef unsigned I = arr.shape[0], i=0
    cdef unsigned J = arr.shape[1], j=0
    cdef unsigned K = arr.shape[2], k=0
    cdef int total = 0
    for i in range(I):
        for j in range(J):
            for k in range(K):
                total += arr[i, j, k]
    return total
```

```python
myarray = np.random.randint(0,1e2,size=(100,100,100),dtype=np.int32)
```

```python
%%timeit -n5 -r1
py_total(myarray)
```

```python
%%timeit -n5 -r1
cy1_total(myarray)
```

```python
%%timeit -n100 -r20
cy2_total(myarray)
```

```python
%%timeit -n100 -r20
cy3_total(myarray)
```

```python
%%timeit -n100 -r20
np.sum(myarray)
```
